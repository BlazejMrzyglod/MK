```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definicja funkcji sigmoidalnej zgodnie ze wzorem z dokumentu
def sigmoid(z):
    """
    Oblicza warto funkcji sigmoidalnej dla zadanego wejcia z.
    Wzr: 1 / (1 + e^-z)
    """
    return 1 / (1 + np.exp(-z))

# 2. Obliczenie gradientu (pochodnej)
def sigmoid_derivative(z):
    """
    Oblicza pochodn funkcji sigmoidalnej.
    Wzr: sigma(z) * (1 - sigma(z))
    """
    sig = sigmoid(z)
    return sig * (1 - sig)

# 3. Przygotowanie danych do wykresu
# Generujemy zakres wartoci z (o X) np. od -10 do 10
z = np.linspace(-10, 10, 200)

# Obliczamy wartoci funkcji i jej gradientu
y = sigmoid(z)
y_prime = sigmoid_derivative(z)

# 4. Wywietlenie funkcji wraz z gradientem na jednym wykresie
plt.figure(figsize=(10, 6))

# Wykres funkcji sigmoidalnej
plt.plot(z, y, label=r'Funkcja aktywacji $\sigma(z)$', color='blue', linewidth=2)

# Wykres pochodnej
plt.plot(z, y_prime, label=r'Gradient $\sigma\'(z)$', color='red', linestyle='--', linewidth=2)

# Dodatki estetyczne wykresu
plt.title('Wariant 1: Funkcja Sigmoidalna i jej gradient')
plt.xlabel('Warto wejciowa (z)')
plt.ylabel('Warto wyjciowa')
plt.axhline(0, color='black', linewidth=0.5, linestyle=':')
plt.axvline(0, color='black', linewidth=0.5, linestyle=':')
plt.grid(True, alpha=0.3)
plt.legend()

# Wywietlenie
plt.show()
```
