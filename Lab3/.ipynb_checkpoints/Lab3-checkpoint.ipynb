```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Ładowanie danych
# Plik war7.csv powinien zawierać kolumny: x1, x2, y
data = pd.read_csv('war7.csv', sep=';', decimal=',')
print(data)
x1 = data['x1'].values
x2 = data['x2'].values
y = data['y'].values

# 2. Budowa macierzy A
A = np.column_stack((x1, x2))

# 3. Rozkład SVD (wersja ekonomiczna)
U, S, VT = np.linalg.svd(A, full_matrices=False)

# 4. Obliczenie współczynników metodą SVD
# x = V * inv(S) * UT * y
S_inv = np.diag(1/S)
coeffs = VT.T @ S_inv @ U.T @ y

a, b = coeffs
print(f"Wyznaczone współczynniki: a = {a:.4f}, b = {b:.4f}")

# 5. Weryfikacja alternatywną metodą (pinv)
coeffs_alt = np.linalg.pinv(A) @ y
print(f"Weryfikacja (pinv): a = {coeffs_alt[0]:.4f}, b = {coeffs_alt[1]:.4f}")

# 6. Wizualizacja dopasowania
plt.figure(figsize=(10, 6))
plt.plot(y, 'k-', label='Dane oryginalne (y)')
plt.plot(A @ coeffs, 'r--', label='Regresja (a*x1 + b*x2)')
plt.legend()
plt.title('Porównanie danych rzeczywistych z modelem regresji')
plt.show()
```
